```python
#Creating Group By

groupingobject = school_data_complete.groupby(['school_name','type'])

#Use GroupedByCount to total students

totalstudentsbyschool = groupingobject.count()
totalstudentsbyschool
```

```python
#Counting students by group with math above 70

mathgroupingobject = mathabove70.groupby(['school_name'])
readinggroupobject = readingabove70.groupby(['school_name'])
studentspassingmathbyschool = mathgroupingobject.count()
studentspassingreadingbyschool = readinggroupobject.count()
studentspassingmathbyschool
studentspassingreadingbyschool
percentagepassingmathbyschool = 100*(studentspassingmathbyschool / groupingobject.count())
percentagepassingmathbyschool = percentagepassingmathbyschool.rename(columns={'math_score': 'percent_passing_math'})
percentagepassingmathbyschool.head()
percentagepassingreadingbyschool = 100*(studentspassingreadingbyschool / groupingobject.count())
percentagepassingreadingbyschool = percentagepassingreadingbyschool.rename(columns={'reading_score':"percent_passing_reading"
                                                                                   })
percentagepassingreadingbyschool = percentagepassingreadingbyschool[["percent_passing_reading"]]
percentagepassingreadingbyschool.head()
percentagepassingmathbyschool = percentagepassingmathbyschool[["percent_passing_math"]]
percentagepassingmathbyschool.head()
```
