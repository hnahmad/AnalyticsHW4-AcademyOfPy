```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#first get a list of unique schools
uniqueschools = school_data_complete["school_name"].unique()
uniqueschools
```

```python
numberofschools = school_data_complete["school_name"].nunique()
numberofschools
```

```python
numberofstudents = school_data_complete["student_name"].count()
numberofstudents
```

```python
#Easier to work with a dataframe of just school and budget

schoolandbudgetdf = school_data_complete[["school_name","budget"]]


                                 
                                       
```

```python
schoolandbudgetdf2 = schoolandbudgetdf.drop_duplicates('school_name')
schoolandbudgetdf2
```

```python
schoolbudgettotal = schoolandbudgetdf2['budget'].sum()
schoolbudgettotal
```

```python
meanmath = school_data_complete["math_score"].mean()
meanmath
```

```python
meanread = school_data_complete["reading_score"].mean()
meanread
```

```python
#Calculates and Appends Average Score per Studnet into DataFrame

average_score = school_data_complete["math_score"]/2 + school_data_complete["reading_score"]/2
school_data_complete['average_score']=average_score
school_data_complete.head()
```

```python
#Calculates Overage Average
overall_average_score = school_data_complete['average_score'].mean()
overall_average_score
```

```python
#This is an alternative way to get same thing
alternativemethod = (school_data_complete['math_score'].mean() + school_data_complete['reading_score'].mean())/2
alternativemethod
```

```python
mathabove70 = school_data_complete.loc[school_data_complete['math_score']>=70,:]
mathabove70.head(20)
```

```python
readingabove70 = school_data_complete.loc[school_data_complete['reading_score']>=70,:]
readingabove70.head(20)
```

```python
mathabove70['student_name'].count()
```

```python
readingabove70['student_name'].count()
```

```python
mathabove70percentage = (mathabove70['student_name'].count() / school_data_complete['student_name'].count())*100
mathabove70percentage
```

```python
readingabove70percentage = (readingabove70['student_name'].count() / school_data_complete['student_name'].count())*100
readingabove70percentage
```

```python
displaydf=pd.DataFrame(
        {"Total Schools": [numberofschools],
          "Total Students": [numberofstudents],
          "Total Budget": [schoolbudgettotal],
          "Avg. Math Score": [meanmath],
          "Avg. Reading Score": [meanread],
          "Avg. Overall Score": [overall_average_score],
          "Math Above 70 (%)": [mathabove70percentage],
          "Reading Above 70(%)": [readingabove70percentage]}
)
displaydf
```
